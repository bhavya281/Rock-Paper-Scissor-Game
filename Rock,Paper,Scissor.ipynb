```python
# Import Required Library
from tkinter import *
import random

root = Tk()
root.geometry("500x450")
root.title("Rock, Paper, Scissor Game")

comp_score=0
player_score=0
options=["Rock","Paper","Scissor"]

def reset_game():
    b1["state"] = "active"
    b2["state"] = "active"
    b3["state"] = "active"
    l2.config(text = "")

def button_disable():
    b1["state"] = "disable"
    b2["state"] = "disable"
    b3["state"] = "disable"

def end_game():
    b1["state"] = "active"
    b2["state"] = "active"
    b3["state"] = "active"
    l2.config(text="")
    l5.config(text="")
    l6.config(text="")
    global player_score
    global comp_score
    player_score=0
    comp_score=0

def rock():
    global player_score
    global comp_score
    computer=random.choice(options)
    if computer== "Rock":
        match_result = "Tie!!!"
    elif computer=="Scissor":
        match_result = "You Win!!! Rock destroys Scissor"
        player_score+=1
    else:
        match_result = "You Lose! Paper covers Rock"
        comp_score+=1
    l2.config(text = match_result)
    l5.config(text = player_score)
    l6.config(text = comp_score)
    button_disable()

def paper():
    global player_score
    global comp_score
    computer=random.choice(options)
    if computer== "Paper":
        match_result = "Tie!!!"
    elif computer=="Scissor":
        match_result = "You Lose! Scissor cuts Paper"
        comp_score+=1
    else:
        match_result = "You Win!!! Paper covers Rock"
        player_score+=1
    l2.config(text = match_result)
    l5.config(text = player_score)
    l6.config(text = comp_score)
    button_disable()

def scissor():
    global player_score
    global comp_score
    computer=random.choice(options)
    if computer == "Scissor":
        match_result = "Tie!!!"
    elif computer == "Rock":
        match_result = "You Lose! Rock destroys Scissor"
        comp_score+=1
    else:
        match_result = "You Win!!! Scissor cuts Paper"
        player_score+=1
    l2.config(text = match_result)
    l5.config(text = player_score)
    l6.config(text = comp_score)
    button_disable()  
    
Label(root,text = "Rock Paper Scissor", font = "normal 17 bold", fg = "red").pack(pady = 20)

frame = Frame(root)
frame.pack()

l1 = Label(frame,text = "Player VS Computer ",font = 13)
l1.pack()
l2 = Label(root,text = "",font = "normal 15",bg = "white",width = 40 ,borderwidth = 2,relief = "solid")
l2.pack(pady = 20)

frame1 = Frame(root)
frame1.pack()
l3 = Label(frame1,text = "Player Score",font = 13)
l3.pack(side=LEFT, padx=20,pady=20)
l4 = Label(frame1,text = "Computer Score",font = 13)
l4.pack(padx=20,pady=20)

frame3 = Frame(root)
frame3.pack()
l5= Label(frame3,text = "",font = "normal 12",bg = "white",width = 10 ,borderwidth = 2,relief = "solid")
l5.pack(side=LEFT,padx=25,pady=10)
l6= Label(frame3,text = "",font = "normal 12",bg = "white",width = 10 ,borderwidth = 2,relief = "solid")
l6.pack(side=LEFT,padx=25,pady=10)

frame2 = Frame(root)
frame2.pack()
b1 = Button(frame2, text = "Rock",font = 10, width = 7,command = rock)
b2 = Button(frame2, text = "Paper ",font = 10, width = 7,command = paper)
b3 = Button(frame2, text = "Scissor",font = 10, width = 7,command = scissor)
b4 = Button(root, text = "Reset",font = 10, width = 7,fg="red",command = reset_game)
b5 = Button(root, text = "End",font = 10, width = 7,fg="red",command = end_game)

b1.pack(side = LEFT, padx = 10,pady=10)
b2.pack(side = LEFT,padx = 10,pady=10)
b3.pack(padx = 10,pady=10)
b4.pack(padx=10,pady=10)
b5.pack(padx=10,pady=10)
root.mainloop()
```
